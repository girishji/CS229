```python
# Plot sigmoid function. Is this a straight line?

min = -100
max = 100
plt.plot(list(range(min, max)), [1 / (1 + math.exp(-x)) for x in range(min, max)]);
```

If we use the criteria like y = 1 if h(x) >= 0, then upper half of the above graph is +ve.

In logistic regression, we have $\theta^Tx$.

Let's plot this:



```python
# theta = [theta_0, theta_1] 
min = -10
max = 10
theta = np.array([3, -5])
plt.plot(list(range(min, max)), [1 / (1 + math.exp(-(theta @ [1, x]))) for x in range(min, max)]);
```

For 1-D of x, the demarcation is a point (where h(x) = 0.5). All x's gets separated as +1 or 0, to either side of this point. For 2-D of x, this demarcation is a stright line. For 3-D of x, this will be a plane.
